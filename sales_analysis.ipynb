### Assignment summary

**Goals**

Understand and advise on:
- Current Sales Forecast globally and by region,
- Quarter‑to‑Date Bookings,
- Coverage,
- Insights and next steps.

**Context**
- Data is at the Product level for Pipeline, Forecast, Bookings, and Target.
- Regions roll up: AMER, EMEA, APAC, JAPAN to Global.
- Direct and Channel are tracked separately.
- Forecast = final expected Net New ARR for the quarter. Includes bookings to date + expected future bookings.
- Pipeline includes only opps already in CRM.
- Opportunity = pipeline at Stage ≥ 2, entering pipeline on Qualified Date.


**KPI definitions**
- Projected Attainment = ```Forecast / Target``` (by Region and Product)
- Remaining Target = ```Target - Bookings```
- Pipeline Coverage = ```Pipeline / Remaining Target```
- QTD Bookings = ```sum of Bookings``` quarter to date

## Findings, decisions and assumptions 

- **Quarter to date meaning**: Given the different dates provided over the spreadsheets, need to define what "QTD" means in this context.

  > **Decision**: Used **max qualified Date** in Pipeline as an anchor. Date goes to April 11th, evaluating then Q2-25.

- **Two columns of “Type”**:  
  • In *Bookings and Pipeline*, it’s **Sales motion** (Net New / Upsell / Renewal).  
  • In *Targets*, it’s **Route to Market** (Direct / Channel / Total).  
  > **Decision**: rename *Bookings/Pipeline* “Type” → `Sales Motion` to avoid confusion.

- **Forecast contains multiple ‘Levels’** (Region, Manager).  
  > **Decision**: use **Level = 'Region'** only, so we don’t double-count.

- **Targets** including GLOBAL and a “Product = Total” line. Merging those with region+product would be a double add.  
  > **Decision**: keep **Type = 'Total'** and **exclude Region='GLOBAL'**

- **Stage is a string**.  
  > **Decision**: parse a numeric `Stage_Num` and define pipeline for coverage as **Stage 2–5** (open, in‑quarter).

## Data cleaning & standardizing

````python

# Rename 'Type' to 'Sales Motion' (Net New / Upsell / Renewal)
if "Type" in df_bookings.columns:
    df_bookings = df_bookings.rename(columns={"Type":"SalesMotion"})
if "Type" in df_pipeline.columns:
    df_pipeline = df_pipeline.rename(columns={"Type":"SalesMotion"})
````

````python
# Numeric coercion for $ fields
for df in [df_bookings, df_pipeline]:
    if "Net New ACV (USD)" in df.columns:
        df["Net New ACV (USD)"] = pd.to_numeric(df["Net New ACV (USD)"], errors="coerce")
````

````python
# Normalizing Region & Product columns for joins
df_bookings["Region"]  = df_bookings.get("CRM Region", df_bookings.get("Region"))
df_pipeline["Region"]  = df_pipeline.get("CRM Region", df_pipeline.get("Region"))
df_bookings["Product"] = df_bookings.get("Product Flag", df_bookings.get("Product"))
df_pipeline["Product"] = df_pipeline.get("Product Flag", df_pipeline.get("Product"))
````

````python
# Parse numeric stage from strings given the data type (like "Stage 3")
def parse_stage(s):
    try:
        if pd.isna(s): return np.nan
        if isinstance(s,(int,float)): return int(s)
        import re
        m = re.search(r"(\d+)", str(s))
        return int(m.group(1)) if m else np.nan
    except Exception:
        return np.nan

df_pipeline["Stage_Num"] = df_pipeline["Stage"].apply(parse_stage)
````

````python
# Targets (Type=Total only), exclude GLOBAL

df_targets_use = df_targets.copy()
df_targets_use = df_targets_use[df_targets_use["Type"].astype(str).str.lower().eq("total")]
df_targets_use = df_targets_use[~df_targets_use["Region"].astype(str).str.upper().eq("GLOBAL")]
df_targets_use = df_targets_use.rename(columns={"Net New ACV Target":"Target"})
````

````python
# Forecast — only Region-level rows for exec math
df_forecast_use = df_forecast.copy()
if "Level" in df_forecast_use.columns:
    df_forecast_use = df_forecast_use[df_forecast_use["Level"].astype(str).str.lower().eq("region")]
df_forecast_use = df_forecast_use.rename(columns={"Forecast Value":"Forecast"})
````

````python
# Define quarter window & snapshot
Q2_START = pd.Timestamp("2025-04-01")
Q2_END   = pd.Timestamp("2025-06-30")
AS_OF    = pd.to_datetime(df_pipeline["Qualified Date"]).max()
print("Snapshot as of:", AS_OF.date())

display(df_targets_use.head(3))
display(df_forecast_use.head(3))
````

````python
# KPI calculation (QTD, Forecast vs Target, Coverage)

# QTD Bookings (in Q2 window up to AS_OF)
bookings_qtd = df_bookings[(df_bookings["Close Date"]>=Q2_START) & (df_bookings["Close Date"]<=AS_OF)].copy()
bookings_qtd["Bookings"] = bookings_qtd["Net New ACV (USD)"]


````

````python
# Defining open in‑quarter pipeline for coverage: Stage 2 to 5, Close Date in Q2, as of Qualified 
pipeline_open_q2 = df_pipeline[
    df_pipeline["Stage_Num"].between(2,5, inclusive="both") &
    (df_pipeline["Qualified Date"]<=AS_OF) &
    (df_pipeline["Close Date"]>=Q2_START) &
    (df_pipeline["Close Date"]<=Q2_END)
].copy()
pipeline_open_q2["Pipeline"] = pipeline_open_q2["Net New ACV (USD)"]

````

````python
# Aggregate  Region and Product

agg_b = bookings_qtd.groupby(["Region","Product"], dropna=False)["Bookings"].sum().reset_index()
agg_p = pipeline_open_q2.groupby(["Region","Product"], dropna=False)["Pipeline"].sum().reset_index()
agg_t = df_targets_use.groupby(["Region","Product"], dropna=False)["Target"].sum().reset_index()
agg_f = df_forecast_use.groupby(["Region","Product"], dropna=False)["Forecast"].sum().reset_index()

````

````python
# Building KPIs table

kpi = agg_t.merge(agg_f, on=["Region","Product"], how="left")\
           .merge(agg_b, on=["Region","Product"], how="left")\
           .merge(agg_p, on=["Region","Product"], how="left")

for c in ["Forecast","Bookings","Pipeline"]:
    kpi[c] = pd.to_numeric(kpi[c], errors="coerce").fillna(0.0)

kpi["Remaining Target"]     = (kpi["Target"] - kpi["Bookings"]).clip(lower=0)
kpi["Projected Attainment"] = np.where(kpi["Target"]>0, kpi["Forecast"]/kpi["Target"], np.nan)
kpi["Coverage"]             = np.where(kpi["Remaining Target"]>0, kpi["Pipeline"]/kpi["Remaining Target"], np.nan)

````

````python
# Region totals

kpi_region = kpi.groupby("Region", as_index=False).agg({
    "Target":"sum","Forecast":"sum","Bookings":"sum","Pipeline":"sum","Remaining Target":"sum"
})
kpi_region["Projected Attainment"] = np.where(kpi_region["Target"]>0, kpi_region["Forecast"]/kpi_region["Target"], np.nan)
kpi_region["Coverage"]             = np.where(kpi_region["Remaining Target"]>0, kpi_region["Pipeline"]/kpi_region["Remaining Target"], np.nan)
kpi_region["QTD % of Target"]      = np.where(kpi_region["Target"]>0, kpi_region["Bookings"]/kpi_region["Target"], np.nan)

````

````python
# Product totals

kpi_product = kpi.groupby("Product", as_index=False).agg({
    "Target":"sum","Forecast":"sum","Bookings":"sum","Pipeline":"sum","Remaining Target":"sum"
})
kpi_product["Projected Attainment"] = np.where(kpi_product["Target"]>0, kpi_product["Forecast"]/kpi_product["Target"], np.nan)
kpi_product["Coverage"]             = np.where(kpi_product["Remaining Target"]>0, kpi_product["Pipeline"]/kpi_product["Remaining Target"], np.nan)

````

## SQL - KPIs 


````python
import sqlite3

con = sqlite3.connect(":memory:")
# Stage minimal slices
bookings_qtd[["Region","Product","Bookings"]].to_sql("bookings_qtd", con, index=False, if_exists="replace")
pipeline_open_q2[["Region","Product","Pipeline"]].to_sql("pipeline_open_q2", con, index=False, if_exists="replace")
df_forecast_use[["Region","Product","Forecast"]].to_sql("forecast", con, index=False, if_exists="replace")
df_targets_use[["Region","Product","Target"]].to_sql("targets", con, index=False, if_exists="replace")

sql_att = '''
SELECT
  t.Region,
  SUM(t.Target)   AS Target,
  SUM(f.Forecast) AS Forecast,
  CASE WHEN SUM(t.Target) > 0 THEN SUM(f.Forecast) * 1.0 / SUM(t.Target) END AS Projected_Attainment
FROM targets t
LEFT JOIN forecast f
  ON t.Region = f.Region AND t.Product = f.Product
GROUP BY 1
ORDER BY 1;
'''
sql_cov = '''
WITH reg AS (
  SELECT
    t.Region,
    SUM(t.Target) AS Target,
    COALESCE(SUM(b.Bookings),0) AS Bookings,
    COALESCE(SUM(p.Pipeline),0) AS Pipeline
  FROM targets t
  LEFT JOIN bookings_qtd b
    ON t.Region = b.Region AND t.Product = b.Product
  LEFT JOIN pipeline_open_q2 p
    ON t.Region = p.Region AND t.Product = p.Product
  GROUP BY 1
)
SELECT
  Region,
  Target,
  Bookings,
  CASE WHEN Target - Bookings > 0 THEN Pipeline * 1.0 / (Target - Bookings) END AS Coverage
FROM reg
ORDER BY 1;
'''

att_df = pd.read_sql(sql_att, con)
cov_df = pd.read_sql(sql_cov, con)
display(att_df)
display(cov_df)
con.close()
````
